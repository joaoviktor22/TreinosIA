## Visualização baseada no posicionamento de pontos utilizando Análise de Componentes Principais

Reduz a dimensionalidade dos dados

```python
raw_data
```

```python
label = raw_data["sex"]
raw_data = raw_data.drop(labels=["sex"],axis=1)
```

```python
dataset = raw_data
dataset = dataset.replace('No',0)
dataset = dataset.replace('Yes',1)
#dataset = dataset.replace('Male',0)
#dataset = dataset.replace('Female',1)
dataset = dataset.replace('Lunch',0)
dataset = dataset.replace('Dinner',1)
dataset = dataset.replace('Sun',0)
dataset = dataset.replace('Sat',1)
dataset = dataset.replace('Thur',0.33)
dataset = dataset.replace('Fri',0.66)
X = dataset.values
y = label.values
a = dataset.columns
#normalize = preprocessing.MinMaxScaler()
#xscaled = normalize.fit_transform(X)
scaler = preprocessing.StandardScaler().fit(X)
xscaled = scaler.transform(X)
dataset = pd.DataFrame(xscaled,columns=a)
dataset
```

# Visualizações baseadas em Projeções Multi-Dimensionais

```python
X = dataset.values
y = label.values

sklearn_pca = sklearnPCA(n_components=2)
X_embedded = sklearn_pca.fit_transform(X)

#pca = PCA(n_components=2)
#components = pca.fit_transform(X)

fig = px.scatter(X_embedded, x=0, y=1, color=label.astype(object))
fig.show()
```

```python
xisomap = manifold.Isomap(10, 2).fit_transform(X)
fig = px.scatter(xisomap, x=0, y=1, color=label.astype(object))
fig.show()
```

```python
scoretestisomap = [(1 , 1 )]
npresisomap = pd.DataFrame(scoretestisomap, columns = ['Score' , 'K'])

for x in range(1, 50, 2):
  saida = manifold.trustworthiness(X, xisomap, n_neighbors=x, metric='euclidean')
  npresisomap=npresisomap.append({'Score' : saida , 'K' : x} , ignore_index=True)
  
npresisomap = npresisomap.drop(0)
fig = px.line(npresisomap, x="K", y="Score", title='NPres do ISOMAP')
fig.show()
```

```python
mds = manifold.MDS(2, max_iter=100, n_init=1)
xmds = mds.fit_transform(X)
fig = px.scatter(xmds, x=0, y=1, color=label.astype(object))
fig.show()
```

```python
scoretestmds = [(1 , 1 )]
npresmds = pd.DataFrame(scoretestmds, columns = ['Score' , 'K'])

for x in range(1, 50, 2):
  saida = manifold.trustworthiness(X, xmds, n_neighbors=x, metric='euclidean')
  npresmds=npresmds.append({'Score' : saida , 'K' : x} , ignore_index=True)
  
npresmds = npresmds.drop(0)
fig = px.line(npresmds, x="K", y="Score", title='NPres do MDS')
fig.show()
```

```python
se = manifold.SpectralEmbedding(n_components=2,
                                n_neighbors=10)
xspec = se.fit_transform(X)
fig = px.scatter(xspec, x=0, y=1, color=label.astype(object))
fig.show()
```

```python
scoretestspec = [(1 , 1 )]
npresspec = pd.DataFrame(scoretestspec, columns = ['Score' , 'K'])

for x in range(1, 50, 2):
  saida = manifold.trustworthiness(X, xspec, n_neighbors=x, metric='euclidean')
  npresspec=npresspec.append({'Score' : saida , 'K' : x} , ignore_index=True)
  
npresspec = npresspec.drop(0)
fig = px.line(npresspec, x="K", y="Score", title='NPres do Spectral Embedding')
fig.show()
```

```python
tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
xtsne = tsne.fit_transform(X)
fig = px.scatter(xtsne, x=0, y=1, color=label.astype(object))
fig.show()
```

```python
scoretesttsne = [(1 , 1 )]
nprestsne = pd.DataFrame(scoretesttsne, columns = ['Score' , 'K'])

for x in range(1, 50, 2):
  saida = manifold.trustworthiness(X, xtsne, n_neighbors=x, metric='euclidean')
  nprestsne=nprestsne.append({'Score' : saida , 'K' : x} , ignore_index=True)
  
nprestsne = nprestsne.drop(0)
fig = px.line(nprestsne, x="K", y="Score", title='NPres do TSNE')
fig.show()
```

```python
nca = NeighborhoodComponentsAnalysis(max_iter=15, random_state=0)
nca = nca.fit(xtsne,y)
xneig = nca.transform(xtsne)
fig = px.scatter(xneig, x=0, y=1, color=label.astype(object))
fig.show()
```
