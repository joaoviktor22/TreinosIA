```
import pandas as pd
import numpy as np
from sklearn import preprocessing

rawData = pd.read_csv("sample_data/students2.csv")
```

```
rawData['time'] = pd.Categorical(rawData['time'])

rawData['time_ord'] =rawData.time.cat.codes

#print(rawData['time_ord'].head())

rawData = rawData.drop(['time'],axis=1)
print(rawData['time_ord'])
```

```
print(rawData.head())
print(rawData.loc[[0][0]])
```

```
rawData = pd.get_dummies(data=rawData, columns=['Gender','Mother_occupation','Father_occupation','coaching','Caste','twelve_education','medium','Class_ten_education'])
```

```
def mudar(Perfomance): 
    if Perfomance == 'Excellent':
        valor = float(1)
    elif Perfomance == 'Vg':
        valor = float(0.66)
    elif Perfomance == 'Good':
        valor = float(0.33)
    else:
        valor = 0
    return valor
```

```
cat_list = []
for i in range(len(rawData)):
    per = rawData.iloc[i]['Performance']
    cat = mudar(per)
    cat_list.append(cat)
rawData['Performance'] = cat_list
```

```
cat_list = []
for i in range(len(rawData)):
    per = rawData.iloc[i]['Class_ X_Percentage']
    cat = mudar(per)
    cat_list.append(cat)
rawData['Class_ X_Percentage'] = cat_list
```

```
cat_list = []
for i in range(len(rawData)):
    per = rawData.iloc[i]['Class_ XII_Percentage']
    cat = mudar(per)
    cat_list.append(cat)
rawData['Class_ XII_Percentage'] = cat_list
```

```
print(rawData.head())
print(rawData.loc[0])
```

```
df_min_max_scaled = rawData.copy() 
column = 'time_ord'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())     
rawData = df_min_max_scaled
```
