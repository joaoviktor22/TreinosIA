```
def mudar(Perfomance): 
    if Perfomance == 'Excellent':
        valor = float(1)
    elif Perfomance == 'Vg':
        valor = float(0.66)
    elif Perfomance == 'Good':
        valor = float(0.33)
    else:
        valor = 0
    return valor
```
